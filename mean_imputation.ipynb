```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from utils import mape, mae, rmse
from data_loaders import load_dataset
from sklearn.impute import SimpleImputer
```

```python
DATASET = 'fun'
metrics = {'rmse':[], 'mae':[], 'mape':[]}

for seed in range(3):
    tr_idxs, tr_vals, te_idxs, te_vals, data_shape = load_dataset(DATASET, seed=seed)
    T, N, D = data_shape
    df_tr_idx = pd.DataFrame(tr_idxs).astype(int)
    df_te_idx = pd.DataFrame(te_idxs).astype(int)

    preds = []
    targets = []
    for i in range(N):
        # init empty firm matrix
        firm_df = np.zeros((T,D))
        firm_df.fill(np.nan)
        # grab one firm's training values, place in firm df
        firm_tr_idx = df_tr_idx[df_tr_idx[1] == i].values
        firm_te_idx = df_te_idx[df_te_idx[1] == i].values

        for (time,firm,feat), val in zip(firm_tr_idx, tr_vals):
            firm_df[time,feat] = val

        print(firm_df)
        print('\n\n\n\n\n\n')
        prev_shape = firm_df.shape
        mean_imputer = SimpleImputer(strategy='mean', missing_values=np.nan)
        firm_df = mean_imputer.fit_transform(firm_df)
        print(firm_df)
        
        for (time,firm,feat), val in zip(firm_te_idx, te_vals):
            preds.append(firm_df[time,feat])
            targets.append(val)

    preds = np.array(preds)
    #preds = preds[~np.isnan(preds)]
    targets = np.array(targets)
    #targets = targets[~np.isnan(preds)]
    metrics['rmse'].append(rmse(targets, preds))
    metrics['mae'].append(mae(targets, preds))
    metrics['mape'].append(mape(targets, preds))

for m, vals in metrics.items():
    print(f'{m}: {np.mean(vals)} +/- {np.std(vals, ddof=1):.6f}')
```
