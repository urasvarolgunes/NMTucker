```python
import pickle
import matplotlib.pyplot as plt
```

```python
def read_pickle(path):
    with open(path, 'rb') as f:
        loaded = pickle.load(f)
    return loaded
```

```python
ML1_dic = read_pickle('/home/uras/Desktop/NMTucker/overfit_curves/ML1_chars_[80, 80, 80]_[10, 10, 10]_[5, 5, 5]_regNone_L1=0.001_L2=0.001_exp0.pkl')
ML2_dic = read_pickle('/home/uras/Desktop/NMTucker/overfit_curves/ML2_chars_[150, 150, 150]_[78, 78, 78]_[5, 5, 5]_regNone_L1=0.001_L2=0.001_exp0.pkl')
#ML3_dic = read_pickle('/home/uras/Desktop/NMTucker/overfit_curves/ML3_chars_[150, 150, 150]_[120, 120, 120]_[75, 75, 75]_regNone_L1=0.001_L2=0.001_exp0.pkl')
#reg_dic = read_pickle('/home/uras/Desktop/NMTucker/overfit_curves/ML1_chars_[80, 80, 80]_[10, 10, 10]_[5, 5, 5]_regL1_L1=1e-07_L2=0.001_exp0.pkl')
```

```python
figure, axis = plt.subplots(1, 2, figsize=(10,5))

axis[0].plot(ML1_dic['train_loss'], label='NMTucker1')
axis[0].plot(ML2_dic['train_loss'], label='NMTucker2')
# plt.plot(ML3_dic['train_loss'], label='NMTucker3')
# plt.plot(reg_dic['train_loss'], label='NMTucker-L1')
axis[0].set_title('Training Loss')
axis[0].set_xlabel('epochs')
axis[0].set_ylabel('MSE')
axis[0].legend()

axis[1].plot(ML1_dic['val_loss'], label='NMTucker1')
axis[1].plot(ML2_dic['val_loss'], label='NMTucker2')
# plt.plot(ML3_dic['val_loss'], label='NMTucker3')
# plt.plot(reg_dic['val_loss'], label='NMTucker-L1')
axis[1].set_title('Validation Loss')
axis[1].set_xlabel('epochs')
axis[1].set_ylabel('MSE')
axis[1].legend()
```
